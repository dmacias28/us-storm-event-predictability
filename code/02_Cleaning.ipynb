### 6.2. begin_date_time & end_date_time

These 2 columns are currently object dtypes and will need to be converted to datetime dtypes. With a Time Series Analysis in mind, the begin_date_time column will eventually become the index, and will need to be sorted from the earliest date to the latest date. However, only the last 2 numbers of the year are listed for each event, making it difficult for the pd.to_datetime() function to distinguish between years that belong to the 1900s and the 2000s. The first 2 numbers of the year will need to be added manually for proper century assignment.

```python
# Sorting the dataframe by year

all_storms = all_storms.sort_values(by='year')
```

```python
# Manually checking for the index where the year 2000 rows start for begin_date_time

print(all_storms['begin_date_time'][0:420_154])
print(all_storms['begin_date_time'][420_154:])
```

```python
# Manually checking for the index where the year 2000 rows start for end_date_time

print(all_storms['end_date_time'][0:420_154])
print(all_storms['end_date_time'][420_154:])
```

```python
# Using the index where the year 2000 rows start (420_154) to specify what rows will need a '19' added to the year for both begin_date_time and end_date_time columns

all_storms['begin_date_time'][0:420_154] = all_storms['begin_date_time'][0:420_154].astype(str).str[:7] + '19' + all_storms['begin_date_time'][0:420_154].astype(str).str[7:]
all_storms['end_date_time'][0:420_154] = all_storms['end_date_time'][0:420_154].astype(str).str[:7] + '19' + all_storms['end_date_time'][0:420_154].astype(str).str[7:]
```

```python
# Using the index where the year 2000 rows start (420_154) to specify what rows will need a '20' added to the year for both begin_date_time and end_date_time columns

all_storms['begin_date_time'][420_154:] = all_storms['begin_date_time'][420_154:].astype(str).str[:7] + '20' + all_storms['begin_date_time'][420_154:].astype(str).str[7:]
all_storms['end_date_time'][420_154:] = all_storms['end_date_time'][420_154:].astype(str).str[:7] + '20' + all_storms['end_date_time'][420_154:].astype(str).str[7:]
```

### 6.3. damage_crops & damage_property

These 2 columns are currently object dtypes but will need to be converted to int dtypes. Their values should be the full dollar amount in damages, not a shorthand value. For example, 5K should be 5,000 and 12.95M should be 12,950,000.

#### 6.3.1. damage_crops

```python
# damage_crops summary

print('damage_crops')
print('Missing Values: ' + str(all_storms['damage_crops'].isnull().sum().sum()))
print('Unique Value Count: ' + str(all_storms['damage_crops'].nunique()))
all_storms['damage_crops'].value_counts()
```

```python
# Filling in all null values with the string '0' since the lack of a value likely means there was no damage

all_storms['damage_crops'] = all_storms['damage_crops'].fillna('0')
```

```python
# Casting all values in the column to strings in order to perform regex

all_storms['damage_crops'] = all_storms['damage_crops'].astype(str)
```

```python
# Getting rid of any '?' characters

all_storms['damage_crops'] = all_storms['damage_crops'].str.replace('?', '', regex=True)
```

```python
# Function that removes any leading zeros for a string value

def no_lead_zeros(string):
    try:
        return str(int(string)) # If a value can immediately be turned into an int, do nothing, and return the string
    except:
        string = re.sub('\A[A-Za-z]*\Z', '0', string) # If a value contains only non-numeric characters, return a '0' as the string
        return re.sub('\A[0 ]*(?=[0-9])', '', string) # If a value starts with a '0', remove as many leading zeros until the real value begins
```

```python
# Applying the no_lead_zeros function to the column to prevent errors in the following cell

all_storms['damage_crops'] = all_storms['damage_crops'].apply(no_lead_zeros)
```

```python
%%time

# Replaces any shorthand dollar amount character with its mathematical notation, evaluates the new string, and returns the value as an int

repl_dict = {'[kK]': '*1e3', '[mM]': '*1e6', '[bB]': '*1e9', '[tT]': '*1e12'}
all_storms['damage_crops'] = all_storms['damage_crops'].replace(repl_dict, regex=True).map(pd.eval).astype(int)
```

#### 6.3.2. damage_property

```python
# damage_property summary

print('damage_property')
print('Missing Values: ' + str(all_storms['damage_property'].isnull().sum().sum()))
print('Unique Value Count: ' + str(all_storms['damage_property'].nunique()))
all_storms['damage_property'].value_counts()
```

```python
# Filling in all null values with the string '0' since the lack of a value likely means there was no damage

all_storms['damage_property'] = all_storms['damage_property'].fillna('0')
```

```python
# Applying the no_lead_zeros function to the column to prevent errors in the following cell

all_storms['damage_property'] = all_storms['damage_property'].apply(no_lead_zeros)
```

```python
%%time

# Replaces any shorthand dollar amount character with its mathematical notation, evaluates the new string, and returns the value as an int

repl_dict = {'[hH]': '*1e2', '[kK]': '*1e3', '[mM]': '*1e6', '[bB]': '*1e9', '[tT]': '*1e12'}
all_storms['damage_property'] = all_storms['damage_property'].replace(repl_dict, regex=True).map(pd.eval).astype(int)
```

## 7. Creating New Columns

### 7.1. region

```python
# Changing strings in the state column to title case

all_storms['state'] = all_storms['state'].str.title()
```

```python
# Creating a for loop that assigns a state its region in a list

West = ['Washington', 'Oregon', 'California', 'Nevada', 'Idaho', 'Montana', 'Wyoming', 'Utah', 'Colorado', 'Arizona', 'New Mexico']
Midwest = ['North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri', 'Wisconsin', 'Illinois', 'Indiana', 'Michigan', 'Ohio']
South = ['Texas', 'Oklahoma', 'Arkansas', 'Louisiana', 'Mississippi', 'Alabama', 'Georgia', 'Florida', 'South Carolina', 'North Carolina', 'Tennessee', 'Kentucky', 'Virginia', 'West Virginia', 'Maryland', 'Delaware', 'District Of Columbia']
Northeast = ['Pennsylvania', 'New Jersey', 'New York', 'Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont']

region = []

for i in all_storms['state']:
    if i in West:
        region.append('West')
    elif i in Midwest:
        region.append('Midwest')
    elif i in South:
        region.append('South')
    elif i in Northeast:
        region.append('Northeast')
    else:
        region.append(i)
```

```python
# Using the list to create a column

all_storms['region'] = region
```

```python
# Moving the column closer to the front of the dataframe next to state

column_to_move = all_storms.pop('region')
all_storms.insert(1, 'region', column_to_move)
```
